```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Ler o CSV

limite_bairros = pd.read_csv('/content/Limite_de_Bairros.csv')
setores_bairros = pd.read_csv('/content/Setores_Bairros.csv')
setores_censit = pd.read_csv('/content/Setores_Censitarios.csv')

```

```python
#filtro das colunas area_plane e nome da tabela limites_bairros

area_plane = limite_bairros[['area_plane', 'nome']]
area_plane
```

```python
# filtrar na tabela setores_censit cd_setor, v0005

cd_setores = setores_censit[['cd_setor', 'v0005']]
cd_setores
```

```python
#filtrar na limite_bairros as colunas nome, area_plane e codbnum

nome_area_bnum = limite_bairros[['nome', 'area_plane', 'codbnum']]
nome_area_bnum
```

```python
# unir as tabelas cd_setores e setores_bairros em cd_setor

merge1 = pd.merge(cd_setores, setores_bairros, on='cd_setor')
merge1

```

```python
# unir merge 1 a nome_area_bnum em codbnum

merge2 = pd.merge(merge1, nome_area_bnum, on='codbnum')
merge2
```

```python
# calcular v0005 agrupado por area_plane

moradores_por_areaplane = merge2.groupby('area_plane')['v0005'].sum().round(2)
moradores_por_areaplane


```

```python
#gráfico de moradores_por_areaplane em seaborn

sns.set_theme(style='darkgrid')
plt.figure(figsize=(10, 6))
sns.set_palette('pastel')
sns.barplot(x=moradores_por_areaplane.index, y=moradores_por_areaplane.values)
plt.title('Moradores por Área de Planejamento')
#com as legendas dos valores dentro das barras
for i, v in enumerate(moradores_por_areaplane.values):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.xlabel('Área de Planejamento')
plt.ylabel('Quantidade Média de Moradores')
plt.savefig('moradores_por_area.png', dpi=300, bbox_inches='tight')
plt.show()



```

```python
# calcular v0005 agrupado por codbnum

moradores_por_bnum = merge2.groupby('codbnum')['v0005'].sum().round(2)
moradores_por_bnum
```

```python
# unir moradores_por_bnum em limite_bairros por codbnum

merge3 = pd.merge(limite_bairros, moradores_por_bnum, on='codbnum')
merge3

# transformando a coluna v0005 em float

merge3['v0005'] = merge3['v0005'].astype(float)
merge3
```
